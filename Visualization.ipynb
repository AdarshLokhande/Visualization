```python
#LINE PLOT 
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(1,11)
y=2*x
plt.plot(x,y)
plt.show()
```

```python
#for add title axis name
plt.plot(x,y,color="black",linestyle=":",linewidth=5)
plt.title("x vs y")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.show()
```

```python
#2 lines in one plot
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1,color="black",linestyle=":",linewidth=2)
plt.plot(x,y2,color="blue",linewidth=2)
plt.title("2 lines plot")
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.grid(False)
plt.show()
```

```python
#subplot method in line plot
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1,color="black",linestyle="-",linewidth=2)
plt.grid(False)
plt.subplot(1,2,2)
plt.plot(x,y2,color="blue",linewidth=2)
plt.grid(False)
plt.show()
```

```python
#PIE-CHART
fruits=["apple","orange","mango","guava"]
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruits)
plt.legend()
plt.show()
```

```python
fruits=["apple","orange","mango","guava"]
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruits,autopct="%0.3f%%",colors=["yellow","grey","blue","black"])
plt.legend()
plt.show()
```

```python
#DOUGHNUT CHART
fruits=["apple","orange","mango","guava"]
quantity=[67,34,100,29]
plt.pie(quantity,labels=fruits,radius=3)
plt.pie([100],colors=["black"],radius=2)
plt.legend()
plt.show()
```
